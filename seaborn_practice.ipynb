```python
crash_df = sns.load_dataset('car_crashes')
#crash_df
```

Distribution Plot

```python
sns.distplot(crash_df.not_distracted, kde=False, bins=25)
```

Join Plot

```python
sns.jointplot(x='speeding', y='alcohol', kind='hex', bins = 50, data = crash_df)
```

KDE plot

```python
sns.kdeplot(crash_df.speeding)
```

Pair plots

```python
#sns.pairplot(crash_df)

tips_df = sns.load_dataset('tips')
#tips_df
```

```python
sns.pairplot(tips_df, hue = 'smoker', palette= 'Blues')
```

Rug plot


```python
sns.rugplot(tips_df.tip)
```

Styling

```python
sns.set_style('ticks')

plt.figure(figsize=(12,6))

sns.set_context('paper', font_scale=1.4)
sns.jointplot(x='speeding', y= 'alcohol', data = crash_df, kind= 'reg')

sns.despine(left=True, bottom=True, top= True)
```

## **Categorical data**

Bar plots

```python
sns.barplot(x='sex', y='total_bill', data = tips_df, estimator=np.cov)
```

Count Plot

```python
sns.countplot(x= 'sex', data= tips_df, hue= 'smoker')
```

 Box Plot

```python
sns.boxenplot(x = 'sex', y = 'total_bill', data = tips_df)
```

```python
sns.boxenplot(x = 'day', y = 'total_bill', hue = 'sex' ,data = tips_df)
plt.legend(loc=0)
```

**violin Plot**

```python
sns.violinplot(x='day', y= 'total_bill', hue = 'sex', data = tips_df, split=True)

plt.legend(loc = 0)
```

Strip Plot

```python
plt.figure(figsize=(8,5))

sns.stripplot(x='day', y='total_bill', data = tips_df, jitter = True, hue = 'sex', dodge=True)

plt.legend(loc = 0)
```

Swarm Plot

```python
sns.violinplot(x = 'day', y = 'total_bill', data = tips_df)
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, color = 'white', jitter=True)
```

Palletes

```python
plt.figure(figsize=(8,5))

sns.set_style('dark')

sns.set_context('poster')
sns.stripplot(x = 'day', y = 'total_bill', data = tips_df, hue = 'sex', dodge=True)

plt.legend(loc = 0)

```

**Matrix Plot**

Heatmaps

```python
plt.figure(figsize=(8,6))

sns.set_context('paper')

df_mx = crash_df.corr()
df_mx

sns.heatmap(df_mx, annot = True, cmap = 'Blues')
```

```python
plt.figure(figsize=(12,8))

flights = sns.load_dataset('flights')
flights = flights.pivot_table(index='month', columns='year', values='passengers')


sns.heatmap(flights, cmap = 'Blues' , linewidths=1)


```
